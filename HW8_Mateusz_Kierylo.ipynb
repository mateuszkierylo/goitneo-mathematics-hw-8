### **TASK 1**

You are considering the possibility of investing in a startup company that is developing a new product in the market. You have two investment options: Option A and Option B.

The probabilities (p) and cash flows (x, in thousands of dollars) are given under different scenarios:

2en.png

Which company, in terms of expected profitability (mathematical expectation), is more interesting with such data?

```python
# The probabilities and cash flows for both options A and B
probabilities_A = [0.05, 0.25, 0.3, 0.4]
cash_flows_A = [0, 25, 40, 80]

probabilities_B = [0.2, 0.25, 0.25, 0.3]
cash_flows_B = [10, 30, 60, 70]

# Function to calculate expected value
def calculate_expected_value(probabilities, cash_flows):
    return sum(p * x for p, x in zip(probabilities, cash_flows))

# Calculating expected values for both options
expected_value_A = calculate_expected_value(probabilities_A, cash_flows_A)
expected_value_B = calculate_expected_value(probabilities_B, cash_flows_B)

print("expected value A:", expected_value_A)

print("\nexpected value B:", expected_value_B)
```

Conclusions: it's better to invest in option A because 50,25k dollars > 40,5k dollars

### **TASK 2**

A consulting company has offered to take additional actions that will change the investment plan of company A. They will reduce the probability of Scenario 1 to 0 and increase the probability of Scenario 2 to 0.3. However, these measures will cost you 10,000 dollars. Is it worth agreeing to such a proposal?

```python
# Updated probabilities for Option A after consulting adjustments
new_probabilities_A = [0, 0.3, 0.3, 0.4]

# Calculate expected value with new probabilities
new_expected_value_A = calculate_expected_value(new_probabilities_A, cash_flows_A)

# Subtract the cost of the consulting measures (10 000 dollars)
new_expected_value_A_adjusted = new_expected_value_A - 10

print("New expected value A adjusted:", new_expected_value_A_adjusted)

print("\nEarlier expected value A:", expected_value_A)
```

Conlusions: The original expected value (50 250 dollars) is higher than the adjusted expected value under the new conditions (41 500 dollars) after considering the cost of the consulting measures. Thus, it is not worth agreeing to the proposal.

### **TASK 4**

You are interested if the results could be manipulated: investigate whether this is feasible. Generate a normally distributed variable and change several indicators so that the series no longer passes the test for normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Generate a normally distributed variable with 100 elements
np.random.seed(60)
normal_data = np.random.normal(loc=0, scale=1, size=100)

# Manipulate the data by adding extreme values to the first five elements
manipulated_data = normal_data.copy()
manipulated_data[:5] += 5  # Adding extreme values to make the distribution non-normal

# Performing Shapiro-Wilk test for normality on both datasets
_, p_value_original = shapiro(normal_data)
_, p_value_manipulated = shapiro(manipulated_data)

# Plotting the original data
plt.figure(figsize=(6, 4))
plt.hist(normal_data, bins=15, alpha=0.7, color='blue')
plt.title(f'Original Data - Shapiro-Wilk Test p={p_value_original:.4f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plotting the manipulated data
plt.figure(figsize=(6, 4))
plt.hist(manipulated_data, bins=15, alpha=0.7, color='red')
plt.title(f'Manipulated Data - Shapiro-Wilk Test p={p_value_manipulated:.4f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

After adding outliers, the p-value drops significantly to 0,0002 strongly suggesting that the data is no longer normally distributed.

The company you decided to invest in is planned to be opened in Ukraine. Therefore, you turned to sociological research. Find at least one indicator that has a strong correlation with the trust indicator (Q57 Trust to people). Construct the distribution graph of this function.

2) Finding another dataset that has a correlation with the first dataset (confidence index Q57). Calculation of the correlation coefficient.

```python
import pandas as pd

# Load the CSV data file to check the contents and structure
data = pd.read_csv("/content/mat_stat.csv", delimiter=';')
data.head()
```

```python
# Now that the data is correctly loaded, extract Q57 and Q61 columns for correlation analysis
q57_q61_data = data[["Q57", "Q61"]]

# Drop rows with missing values in these columns to ensure accurate correlation calculation
q57_q61_data = q57_q61_data.dropna()

# Calculate the correlation coefficient between Q57 and Q61
correlation_coefficient_corrected = q57_q61_data.corr().iloc[0, 1]
print("Correlation coefficient between Q57 and Q61", correlation_coefficient_corrected)
```

3) Calculation of variance and mathematical expectation for the found data set (Q61)

```python
# Calculate the variance and mean (mathematical expectation) for Q61
variance_q61 = q57_q61_data["Q61"].var()
mean_q61 = q57_q61_data["Q61"].mean()

# Printing the results
print(f"For the dataset Q61, the variance is approximately {variance_q61:.3f}")
print(f"and the mathematical expectation (mean) is approximately {mean_q61:.3f}.")
```

4) Create a distribution histogram and a theoretical distribution curve based on mathematical expectations and variance

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Data for Q61
data_q61 = q57_q61_data["Q61"]

# Histogram of the data
plt.hist(data_q61, bins=30, alpha=0.7, label='Q61 Data', color='blue', density=True)

# Theoretical normal distribution curve based on the calculated mean and variance
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean_q61, np.sqrt(variance_q61))
plt.plot(x, p, 'k', linewidth=2, color='red', label='Theoretical Normal Curve')

title = "Histogram and Theoretical Normal Curve for Q61"
plt.title(title)
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

5) Conclusions:

 The histogram and theoretical normal distribution curve suggest that the Q61 data may not be normally distributed. The distribution appears somewhat skewed and does not align perfectly with the theoretical curve, indicating deviations from normality. This could impact analyses that assume normality
